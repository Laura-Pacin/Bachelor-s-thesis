```python
import pandas as pd
import geopandas as gd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
from google.colab import drive
drive.mount("/content/gdrive")
```

```python
data = pd.read_stata("/content/gdrive/My Drive/productores_relac.dta")
```

```python
data
```

```python
url_col = "https://gist.githubusercontent.com/john-guerra/43c7656821069d00dcbc/raw/be6a6e239cd5b5b803c6e7c2ec405b793a9064dd/Colombia.geo.json"

col = gd.read_file(url_col)
col
```

```python
col2 = col.drop(index=[2,27,28,29, 30, 31, 24, 25, 26])
```

```python
col2
```

```python
col2
```

```python
col2 = col2.reset_index()
```

```python
col2
```

```python
data2 = col2.merge(data, how = "left", left_index = True, right_index = True)

data2
```

```python
fig, ax= plt.subplots(figsize=(10,10))

ax.set_xlabel('Longitud') # X axis
ax.set_ylabel('Latitud') # Y axis

ax= col.plot(column='NOMBRE_DEPT', ax=ax, zorder=5, color="white", edgecolor="Black" )

data2.plot("relac_blanco", ax=ax, zorder=50, cmap="BuPu", legend=True)

fig.patch.set_facecolor('xkcd:white')

plt.title("Productores blancos", size=25)
```

```python
fig, ax= plt.subplots(figsize=(10,10))

ax.set_xlabel('Longitud') # X
ax.set_ylabel('Latitud') # Y

ax= col.plot(column='NOMBRE_DEPT', ax=ax, zorder=5, color="white", edgecolor="Black" )

data2.plot("relac_afro", ax=ax, zorder=50, cmap="Greens", legend=True)

fig.patch.set_facecolor('xkcd:white')

plt.title("Productores afrodescendientes", size=25)
```

```python
fig, ax= plt.subplots(figsize=(10,10))

ax.set_xlabel('Longitud') # X
ax.set_ylabel('Latitud') # Y

ax= col.plot(column='NOMBRE_DEPT', ax=ax, zorder=5, color="white", edgecolor="Black" )

data2.plot("relac_indi", ax=ax, zorder=50, cmap="RdPu", legend=True)

fig.patch.set_facecolor('xkcd:white')

plt.title("Productores indígenas", size=25)
```

```python
ok_pafro_pais = pd.read_stata("/content/gdrive/My Drive/ok_pafro_pais.dta")
```

```python
ok_pafro_pais
```

```python
pafro_pais = ok_pafro_pais.drop(index=[1])
```

```python
pafro_pais
```

```python
n = len(pafro_pais.educacion)
x = np.arange(n)
width = 0.25
plt.bar(x - width, pafro_pais.blanco, width=width, label='Blancos')
plt.bar(x, pafro_pais.afrodescendiente, width=width, label='Afrodescendientes')
plt.bar(x + width, pafro_pais.indigena, width=width, label='Indígenas')
plt.xticks(x, pafro_pais.educacion)
plt.legend(loc='best')
plt.show()
```
